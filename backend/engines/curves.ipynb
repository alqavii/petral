```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np


DATA_DIR = Path("../data/processed") 

rolled_raw = pd.read_csv(DATA_DIR / "cl_front_month_raw.csv")
panama_rolled = pd.read_csv(DATA_DIR / "cl_front_month_panama_adjusted.csv")
proportional_rolled = pd.read_csv(DATA_DIR / "cl_front_month_proportional_adjusted.csv")
```

```python
plt.plot(pd.to_datetime(panama_rolled['ts_event']), panama_rolled['close'], label='CL Front Month Close Prices (Panama Adjusted)', linewidth=0.5)
plt.title('WTI Crude Front Month Close Prices (Panama Adjusted)')
plt.xlabel('Time Index')
plt.ylabel('Close Price (USD)')
fig = plt.gcf()
fig.set_size_inches(12, 6)
fig.set_dpi(600)
plt.tight_layout()
plt.show()
```

```python
plt.plot(pd.to_datetime(rolled_raw['ts_event']), rolled_raw['close'], label='CL Front Month Close Prices (Raw Data)', linewidth=0.5)
plt.title('WTI Crude Front Month Close Prices (Raw Data)')
plt.xlabel('Time Index')
plt.ylabel('Close Price (USD)')
fig = plt.gcf()
fig.set_size_inches(12, 6)
fig.set_dpi(600)
plt.tight_layout()
plt.show()
```

```python
#log_returns = np.log(panama_rolled['close'] / panama_rolled['close'].shift(1)).fillna(0) # type: ignore
arithmetic_returns = panama_rolled["close"].diff().fillna(0)  # type: ignore
```

```python
plt.plot(pd.to_datetime(panama_rolled['ts_event']), arithmetic_returns, label='Arithmetic Returns of CL Front Month Close Prices (Panama Adjusted)', linewidth=0.5)
plt.title('Arithmetic Returns of WTI Crude Front Month Close Prices (Panama Adjusted)')
plt.xlabel('Time Index')
plt.ylabel('Arithmetic Return')
fig = plt.gcf()
fig.set_size_inches(12, 6)
fig.set_dpi(600)
plt.tight_layout()
plt.show()
```
