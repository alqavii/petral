```python
import pandas as pd
import numpy as np
from pathlib import Path


DATA_DIR = Path("../data/cl/raw")

```

```python
files = list(DATA_DIR.glob("*.csv.zst"))
print(files)
```

```python
df = pd.read_csv(files[0])
```

```python
month_codes = "FGHJKMNQUVXZ"
pattern = rf'^CL[{month_codes}]\d$'
#pattern = rf'CL[{month_codes}]\d(?:-?CL[{month_codes}]\d)*$'
#pattern = rf'CL[{month_codes}]\d-CL[{month_codes}]\d$'
cl = df[df['symbol'].str.match(pattern)].copy()
#cl = cl.sort_values(by=['ts_event', 'symbol']).reset_index(drop=True)
cl[["ts_event", "symbol", "close", "volume"]]
```

```python
pattern = rf'CL[{month_codes}]\d-CL[{month_codes}]\d$'
spreads = df[df['symbol'].str.match(pattern)].copy()
spreads[["ts_event", "symbol", "close", "volume"]]
spreads.to_csv(DATA_DIR.parent / "processed" / "cl_spreads.csv", index=False)
```
